```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
import statsmodels.formula.api as smf
import statsmodels.api as sm
from scipy import stats
```

```python
dfdata = pd.read_csv("data.csv")
```

```python
dfsob = pd.get_dummies(dfdata['bplg'], prefix="sob", prefix_sep='')
```

```python
del dfsob['sob56']
```

```python
dfyobdummy = pd.get_dummies(dfdata['yob']-1900, prefix="yob", prefix_sep='')
```

```python
del dfyobdummy['yob54']
```

```python
frame = [dfdata, dfsob, dfyobdummy]
dfdatav2 = pd.concat(frame, axis=1)
```

```python
dfdatav2.head()
```

```python
census70 = np.where(dfdatav2["year"] == 1970, 1, 0)
```

```python
census80 = np.where(dfdatav2["year"] == 1980, 1, 0)
```

```python
male = np.where(dfdatav2["sex"] == 1, 1, 0)
```

```python
race = np.where(dfdatav2["race"] == 1, 1, 0)
```

```python
age2 = dfdatav2['age']^2
```

```python
age3 = dfdatav2['age']^3
```

```python
age4 =dfdatav2['age']^4
```

```python
ny7 = np.where(dfdatav2["numlawyears"] == 7, 1, 0)
```

```python
ny8 = np.where(dfdatav2["numlawyears"] == 8, 1, 0)
```

```python
ny9 = np.where(dfdatav2["numlawyears"] >= 9, 1, 0)
```

```python
controls = ['census80', 'male', 'race', 'age2', 'age3', 'age4', *dfsob, *dfyobdummy]
```

```python
string_controls = 'census70'
for i in controls:
    string_controls = string_controls + '+' + i
```

```python
ols_model = smf.ols('lnwkwage~ gradcap+age+' + string_controls, dfdatav2).fit(cov_type='cluster', cov_kwds={'groups': dfdatav2['yob_bplg']}, sample_weight=dfdatav2['slwt'])
```

```python
ols_model.summary()
```

```python
first_stage = smf.ols("gradcap ~ ny7+ny8+ny9+age+" +string_controls , dfdatav2).fit(cov_type='cluster', cov_kwds={'groups': dfdatav2['yob_bplg']},sample_weight=dfdatav2['slwt'])
first_stage.summary()
```

```python
f_test = first_stage.f_test("ny7=ny8=ny9=0")
print(f_test)
```

```python
year_predicted = first_stage.predict()
```

```python
second_stage = smf.ols("lnwkwage ~ year_predicted+age+" +string_controls, dfdatav2).fit(cov_type='cluster', cov_kwds={'groups': dfdatav2['yob_bplg']}, sample_weight=dfdatav2['slwt'])
second_stage.summary()
```
